```python
import pandas as pd                          
import matplotlib.pyplot as plt 
import numpy as np
import cmocean 

%matplotlib inline                           
%config InlineBackend.figure_format='retina' 
plt.rcParams['figure.figsize'] = 12,6   
```

```python
berkely = pd.read_csv('data/temp/gmt_Berkeley_Earth.csv')
era5 = pd.read_csv('data/temp/gmt_ERA5.csv')
gistemp =pd.read_csv('data/temp/gmt_GISTEMP.csv')
hadcrut5 =pd.read_csv('data/temp/gmt_HadCRUT5.csv')
jra55 =pd.read_csv('data/temp/gmt_JRA-55.csv')
noaa =pd.read_csv('data/temp/gmt_NOAAGlobalTemp.csv')
```

```python

fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(nrows = 6, ncols = 1, figsize = (10,10))

### add the GISS temp dataset to the first plot (ax1):
ax1.plot(berkely['Year'],
        berkely['Berkeley Earth (degC)'],
        linewidth=2,label='Berkely Temp',
        color='black')

### add the hadcrut5 dataset on top and change our label and color!
ax2.plot(hadcrut5['Year'],
        hadcrut5['HadCRUT5 (degC)'],
        linewidth=2,label='HadCRUT5',
        color='red')

ax3.plot(gistemp['Year'],
        gistemp['GISTEMP (degC)'],
        linewidth=2,label='GISTEMP',
        color='blue')
### add the hadcrut5 dataset on top and change our label and color!
ax4.plot(jra55['Year'],
        jra55['JRA-55 (degC)'],
        linewidth=2,label='JRA-55',
        color='green')

ax5.plot(noaa['Year'],
        noaa['NOAAGlobalTemp (degC)'],
        linewidth=2,label='NOAA',
        color='orange')

ax6.plot(era5['Year'],
        era5['ERA5 (degC)'],
        linewidth=2,label='ERA-5',
        color='pink')

#### set up all our labels etc:

## plot a legend on the ax
ax1.legend() 
ax1.axhline(0,color='k')
ax1.set_ylabel('Temp anomaly (deg C)')
ax1.set_xlabel('Year')
ax1.set_xlim(1850,2023) ## set our x-axis limits so that they show the same period for every plot.

## plot a legend on the ax
ax2.legend() 
ax2.axhline(0,color='k')
ax2.set_ylabel('Temp anomaly (deg C)')
ax2.set_xlabel('Year')
ax2.set_xlim(1850,2023) ## set our x-axis limits so that they show the same period for every plot.


## plot a legend on the ax
ax3.legend() 
ax3.axhline(0,color='k')
ax3.set_ylabel('Temp anomaly (deg C)')
ax3.set_xlabel('Year')
ax3.set_xlim(1850,2023) ## set our x-axis limits so that they show the same period for every plot.

## plot a legend on the ax
ax4.legend() 
ax4.axhline(0,color='k')
ax4.set_ylabel('Temp anomaly (deg C)')
ax4.set_xlabel('Year')
ax4.set_xlim(1850,2023) ## set our x-axis limits so that they show the same period for every plot.

## plot a legend on the ax
ax5.legend() 
ax5.axhline(0,color='k')
ax5.set_ylabel('Temp anomaly (deg C)')
ax5.set_xlabel('Year')
ax5.set_xlim(1850,2023) ## set our x-axis limits so that they show the same period for every plot.


## plot a legend on the ax
ax6.legend() 
ax6.axhline(0,color='k')
ax6.set_ylabel('Temp anomaly (deg C)')
ax6.set_xlabel('Year')
ax6.set_xlim(1850,2023) ## set our x-axis limits so that they show the same period for every plot.

##overall figure title
fig.suptitle(' Global Ave. Temperature anomaly Relative to 1850 - 1900 from various databases', fontsize=12)


plt.tight_layout()
```

global temperature anomaly relative to 1850 - 1900 from six different databases compared side by side. Noaa, Hadcrut5, and Berkely have the most extensive datasets, going back to 1850. 

```python
fig, ax = plt.subplots(figsize = (10, 5))

### add the GISS temp dataset to the plot first:
ax.plot(gistemp['Year'],
        gistemp['GISTEMP (degC)'],
        linewidth=2,label='GISTEMP global mean surface air temp anomaly',
        color='yellow')

### add the hadcrut5 dataset on top and change our label and color!
ax.plot(hadcrut5['Year'],
        hadcrut5 ['HadCRUT5 (degC)'],
        linewidth=2,label='HadCRUT5 global mean surface air temp anomaly',
        color='red')

### add the JRA-55 dataset on top and change our label and color!
ax.plot(jra55['Year'],
        jra55['JRA-55 (degC)'],
        linewidth=2,label='JRA-55 global mean surface air temp anomaly',
        color='orange')

ax.plot(noaa['Year'],
        noaa['NOAAGlobalTemp (degC)'],
        linewidth=2,label='NOAA global mean surface air temp anomaly',
        color='green')

ax.plot(era5['Year'],
        era5['ERA5 (degC)'],
        linewidth=2,label='ERA-5 global mean surface air temp anomaly',
        color='pink')

ax.plot(berkely['Year'],
        berkely['Berkeley Earth (degC)'],
        linewidth=2,label='Berkely global mean surface air temp anomaly',
        color='blue')
## plot a legend on the ax
ax.legend() 

## add a horizontal line at 0 !
ax.axhline(0,color='k')

## add plot title:
ax.set_title('Global Ave. Temperature anomaly (deg C) Relative to 1850 - 1900')

## add axes labes:
ax.set_ylabel('Temperataure anomaly (deg C)')
ax.set_xlabel('Year')
```

global average temperature anomalies compared on one plot to better visualize data discrepencies

```python
## Create your figure and axis objects using the subplots() 
## function from pyplot (which you imported using the alias plt):

fig, ax = plt.subplots(figsize=(10,5))

ax.fill_between(berkely['Year'], 0, berkely['Berkeley Earth (degC)'],
             color = "white",
             alpha = 1.0, ### alpha is the transparency!
             label = 'Berkely global mean surface air temp anomaly')

## Set properties of title, ylabels and xlabels
ax.set_title('Mean Surface Air Temperature Anolomalies from Berkely Earth',
             fontsize=18, fontweight='bold')
ax.set_ylabel('Temperature deg C',
             fontsize=12)
ax.set_xlabel('Year',
             fontsize=12)

plt.legend() ## Add legend, with label corresponding the label you set in the plt.plot() above.
plt.grid(linestyle='--') ## Add dashed grid lines

ax.axhline(0,color='k')

### adjust ticklabel font properties:
for tick in ax.get_xticklabels():
    tick.set_fontsize(8)
 
for tick in ax.get_yticklabels():
    tick.set_fontsize(8)
  

ax.set_facecolor("pink")
```

mean surface air temperature anomalies from Berkely Earth from 1850-2023. greater anomalies start to be observed in the early 1900's, increasing rapidly after 1975.
